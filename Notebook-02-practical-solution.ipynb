#### Task 1 

```r
measles <- read.csv("Data/measles.csv")
head(measles)
str(measles)
```

```r
par(mfrow=c(2,2))
hist(measles$Population, breaks=10)
hist(measles$Susceptibles.input, breaks=10)
hist(measles$Measles.reporting.rate, breaks=10)
hist(measles$Percentage.months.measles, breaks=10)
```

#### Task 2

```r
boxplot(measles$Population[measles$Region=="Atlantic"],
measles$Population[measles$Region=="Pacific"], ylab="Population size", xlab=c("Atlantic", "Pacific"))

boxplot(formula=Population~Region, data=measles)
```

#### Task 3

```r
# scatter plot
par(mfrow=c(1,2))
plot(measles$Percentage.months.measles, measles$Susceptibles.input)
plot(measles$Percentage.months.measles, measles$Population)
```
