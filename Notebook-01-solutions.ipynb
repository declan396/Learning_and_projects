### Exercise 01-01-set-theory

A pain assessment chart defines self-perceived levels of pain as discrete integer numbers from 0 (no pain) to 10 (worst possible pain).

Define the sample space of pain level. Define the event of "more than severe pain" with pain level greater than 6.

Assume that out of 40 patients in A&E, 75\% of them are reported as "more than severe pain" while 10\% of them display a pain level of 0. Plot a histogram of the distribution of pain levels at A&E with the information you have.


```r
S <- seq(0,10, 1)
severe <- c(7,8,9,10)
severe %in% S
```

```r
nr_severe <- 40*0.75
nr_zero <- 40*0.10
nr_else <- 40 - nr_severe - nr_zero
```

```r
barplot(c(nr_zero, nr_else, nr_severe), names.arg = c("no pain","not severe","severe"), ylab = "Patients in A&E", main = "Pain assessement chart")

```

###  Exercise 01-02-set-operations

The sample space is $S=\{A,C,G,T\}$ and some possible events are $A=\{G,C,A\}$ and $B=\{T,G\}$. 

What is the complement of the union of $A$ and $B$?

```r
# define the sample space
S <- c("A","C","G","T")
# define A and B
A <- c("G","C","A")
B <- c("T","G")

# complement of A U B
print(setdiff(S,union(A,B)))
```
