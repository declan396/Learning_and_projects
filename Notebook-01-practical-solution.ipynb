#### Task 1

Consider a sample space consisting of all possible amino acids. Do the sets of amino acids with __polar uncharged side chain__ and amino acids __with G as first base__ form a partition of the sample space? Use __R__.

```r
# amino acids
S <- c("F", "L", "V", "M", "I", "S", "P", "T", "A", "Y", "H", "Q", "N", "K", "D", "E", "C", "W", "R", "G")

# amino acids with polar uncharged side chain
A <- c("S", "T", "N", "Q")

# amino acids with G as first base
B <- c("V", "A", "D", "E", "G")
```

```r
# 1) are they disjoint?
length(intersect(A, B))==0

# 2) do they form a partition?
length(setdiff(S, union(A, B)))==0
```

#### Task 2

Given the sample space and events defined in Task 1, use __R__ to prove the set properties of commutativity and associativity.

```r
# commutativity
setequal(union(A,B), union(B,A))

setequal(intersect(A,B), intersect(B,A))
```

```r
C <- c()

# associativity
setequal(union(A,union(B,C)), union(union(A,B),C))

setequal(intersect(A,intersect(B,C)), intersect(intersect(A,B),C))
```

#### Task 4

Perform a sliding-window analysis on the protein by recording the occurrences (of frequencies) of amino acid with polar uncharged side chain over the Collagen protein sequence.

In a sliding-window analysis, you are requested to scan the sequence over windows of fixed length and record (and plot) the metric of interest. Windows can be overlapping or not.

You are free to decide on the windows' length and step (how much the move from one window to the next one), the latter being relevant only for overlapping windows. It is easier to implement non-overlapping windows.

```r
len_protein <- nchar(protein)

len_win <- 100

starts <- seq(1, len_protein, len_win)

starts
```

```r
values <- c()

for (start in starts) {
    
    end <- start + len_win -1
    
    window <- strsplit(protein, split="")[[1]][start:end]
    
    count <- length(which(window %in%A ))
      
    len_window <- length(which(!is.na(window)))
    
    values <- c(values, count / len_window)
    
}
```

```r
plot(values, xlab="Window", type="l")
```
